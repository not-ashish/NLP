```python
import nltk
nltk.download('wordnet')
nltk.download('wordnet_ic')
from nltk.corpus import wordnet_ic

import numpy as np
```

```python
from nltk.corpus import wordnet as wn
pairs=[('the','DT'), ('man','NN'), ('swim','VB'), ('with', 'PR'), ('a', 'DT'),
('girl','NN'), ('and', 'CC'), ('a', 'DT'), ('boy', 'NN'), ('whilst', 'PR'),
('the', 'DT'), ('woman', 'NN'), ('walk', 'VB'),('dog','NN'),('table','NN'),('lamp','NN')]
synset=[]

for pair in pairs:
  try:
    synset.append(wn.synsets(pair[0],pair[1].lower()[0])[0])
    print("\nMost frequent WordNet synset of",pair[0])
    print(wn.synsets(pair[0],pair[1].lower()[0])[0])
  except:
    pass
```

```python
brown_ic=wordnet_ic.ic('ic-brown.dat')
pSimA=[round(synset[0].path_similarity(synset[0]),3)]
wSimA=[round(synset[0].wup_similarity(synset[0]),3)]
liSimA=[round(synset[0].lin_similarity(synset[0],brown_ic),3)]
lSimA=[round(synset[0].lch_similarity(synset[0]),3)]

for i,pair in enumerate(synset):
    for j,pair2 in enumerate(synset[i+1:]):
        print("\nPath Similarity:")
        try:
          pSim=round(pair.path_similarity(pair2),3)
          pSimA.append(round(pair.path_similarity(pair2),3))
          print(str(pair)+" "+str(pair2))
          print(pSim)
        except:
          pSimA.append(0)
          print("None")

        print("\nLeacock Chodorow Similarity:")
        try:
            lSim=round(pair.lch_similarity(pair2)/lSimA[0],3)
            lSimA.append(round(pair.lch_similarity(pair2)/lSimA[0],3))
            print(str(pair)+" "+str(pair2))
            print(lSim)
        except:
            lSimA.append(0)
            print("None")

        print("\nWu-Palmer Similarity:")
        try:
          wSim=round(pair.wup_similarity(pair2),3)
          wSimA.append(round(pair.wup_similarity(pair2),3))
          print(str(pair)+" "+str(pair2))
          print(wSim)
        except:
            wSimA.append(0)
            print("None")

        print("\nlin Similarity:")
        try:
            liSim=round(pair.lin_similarity(pair2,brown_ic),3)
            liSimA.append(round(pair.lin_similarity(pair2,brown_ic),3))
            print(str(pair)+" "+str(pair2))
            print(pSim)
        except:
            liSimA.append(0)
            print("None")
        
        print("Least common subsumer")
        print(pair.lowest_common_hypernyms(pair2))

        print("\n\n")

print ("Normalized Vectors:")
print([pSimA,wSimA,liSimA,lSimA])
```

```python
size = len(synset)

subsumerMat =[[None]*size for _ in range(size)]
pSimMat=np.ndarray((size,size))
wSimMat=np.ndarray((size,size))
lSimMat=np.ndarray((size,size))
liSimMat=np.ndarray((size,size))

for i,pair in enumerate(synset):
    for j,pair2 in enumerate(synset):
        pSim=pair.path_similarity(pair2)
        pSimMat[i,j]=pSim

        try:
            lSim=pair.lch_similarity(pair2)/lSimA[0]
            lSimMat[i,j]=lSim
        except:
            lSimMat[i,j]=0

        wSim=pair.wup_similarity(pair2)
        wSimMat[i,j]=wSim

        try:
            liSim=pair.lin_similarity(pair2,brown_ic)
            liSimMat[i,j]=liSim
        except:
            liSimMat[i,j]=0
        
        subsumers = pair.lowest_common_hypernyms(pair2)
        if len(subsumers) > 0:
            subsumerMat[i][j] = subsumers[0].name()
```

```python
import pandas as pd

word_list = ["man", "swim", "girl", "boy", "woman", "walk", "dog", "table", "lamp"]

pd.set_option('display.float_format', lambda x: '%.3f' % x)

df_subsummer = pd.DataFrame(subsumerMat, index = word_list, columns = word_list)
print("Matrix of least common subsumers")
print(df_subsummer)
```

```python
df_path = pd.DataFrame(pSimMat, index = word_list, columns = word_list)
print("Matrix of Path Similarity Method")
print(df_path)
```

```python
df_wu = pd.DataFrame(wSimMat, index = word_list, columns = word_list)
print("Matrix of Wu-Palmer Similarity Method")
print(df_wu)
```

```python
df_l = pd.DataFrame(lSimMat, index = word_list, columns = word_list)
print("Matrix of Leacock Chodorow Similarity Method")
print(df_l)
```

```python
df_lin = pd.DataFrame(liSimMat, index = word_list, columns = word_list)
print("Matrix of Lin Similarity Method")
print(df_lin)
```
